# Helper functions

```python
def plot_attachment_to_beforedisc_opinion(teams_attachment, eps, fname=''):
    teams_num = len(teams_attachment.keys())
    issues_num = 5
    fig, axes = plt.subplots(teams_num, issues_num, figsize=(issues_num*3, teams_num*3))
    for team_index, team_id in enumerate(sorted(teams_attachment.keys())):
        for issue_index, issue in enumerate(sorted(teams_attachment[team_id].keys())):
            a11 = teams_attachment[team_id][issue]['a11']
            a22 = teams_attachment[team_id][issue]['a22']
            # Plotting the level of attachment to the before discussion opinion dynamics.
            axes[team_index, issue_index].plot(a11, '-x')
            axes[team_index, issue_index].plot(a22, '-p')
            axes[team_index, issue_index].legend(['Person 1', 'Person 2'])
            if team_index == 0:
                axes[team_index, issue_index].set_title(issue)
            if team_index == teams_num - 1:
                axes[team_index, issue_index].set_xlabel('Time')
            if issue_index == 0:
                axes[team_index, issue_index].set_ylabel(
                    'Team {}\'s attachment'.format(team_id))
    if fname != '':
        fig.savefig(fname+'_eps{}.pdf'.format(eps), bbox_inches='tight')
        fig.savefig(fname+'_eps{}.png'.format(eps), bbox_inches='tight')
```

```python
nan_keys = ['0/0', 'n/0', 'xi[k]-xi[k-1]==0', 'wij[k]==0', 'xj[k]-xi[k]==0']

def get_all_aiis(teams_attachment):
    all_aiis = []
    nan_stats = defaultdict(int)
    for team_index, team_id in enumerate(teams_attachment.keys()):
        for issue_index, issue in enumerate(teams_attachment[team_id].keys()):
            a11 = teams_attachment[team_id][issue]['a11']
            a22 = teams_attachment[team_id][issue]['a22']
            all_aiis.extend(a11)
            all_aiis.extend(a22)
            
            a11_nan_details = teams_attachment[team_id][issue]['a11_nan_details']
            a22_nan_details = teams_attachment[team_id][issue]['a22_nan_details']
            
            for i, x in enumerate(a11):
                if len(a11_nan_details[i]) > 0 and not np.isnan(x):
                    print(team_id, issue, 1, i)
            for i, x in enumerate(a22):
                if len(a22_nan_details[i]) > 0 and not np.isnan(x):
                    print(team_id, issue, 2, i)
            
            for nan_key in nan_keys:
                for a11_nan_detail_item in a11_nan_details:
                    if nan_key in a11_nan_detail_item:
                         nan_stats[nan_key] += 1
                for a22_nan_detail_item in a22_nan_details:
                    if nan_key in a22_nan_detail_item:
                        nan_stats[nan_key] += 1
    return np.array(all_aiis), nan_stats
```

# Main  body

```python
n, _ = data.shape
team_ids = np.unique(data['Group'])
issues = ['asbestos', 'disaster', 'school', 'sports', 'surgery']
```

```python
teams_data = defaultdict(dict)

teams_num = 30
issues_num = 5
fig, axes = plt.subplots(teams_num, issues_num, figsize=(issues_num*3, teams_num*3))

for team_index, team_id in enumerate(team_ids):
    this_team_data = data[data['Group'] == team_id]
    for issue_index, issue in enumerate(issues):
        this_team_issue_data = this_team_data[this_team_data['Issue'] == issue]
        this_team_issue_data = this_team_issue_data.sort_values(
            by='Person')   # Just to make sure the first one is person 1.

        p1 = this_team_issue_data.iloc[0]
        x1 = [float(re.sub('\$|,', '', p1['Initial opinion'])),
              float(re.sub('\$|,', '', p1['Period1 opinion'])),
              float(re.sub('\$|,', '', p1['Period2 opinion'])),
              float(re.sub('\$|,', '', p1['Period3 opinion']))]
        w12 = [float(p1['Period1 wij']),
               float(p1['Period2 wij']),
               float(p1['Period3 wij'])]

        p2 = this_team_issue_data.iloc[1]
        x2 = [float(re.sub('\$|,', '', p2['Initial opinion'])),
              float(re.sub('\$|,', '', p2['Period1 opinion'])),
              float(re.sub('\$|,', '', p2['Period2 opinion'])),
              float(re.sub('\$|,', '', p2['Period3 opinion']))]
        w21 = [float(p2['Period1 wij']),
               float(p2['Period2 wij']),
               float(p2['Period3 wij'])]
        teams_data[team_id][issue] = {'x1': x1, 'x2': x2, 'w12': w12, 'w21': w21}

        # Plotting the opinion dynamics.
        axes[team_index, issue_index].plot(x1)
        axes[team_index, issue_index].plot(x2)
        axes[team_index, issue_index].legend(['Person 1', 'Person 2'])
        if team_index == 0:
            axes[team_index, issue_index].set_title(issue)
        if team_index == teams_num - 1:
            axes[team_index, issue_index].set_xlabel('Time')
        if issue_index == 0:
            axes[team_index, issue_index].set_ylabel('Team {}\'s opinions'.format(team_id))

fig.savefig('Figures/Opinions.pdf', bbox_inches='tight')
fig.savefig('Figures/Opinions.png', bbox_inches='tight')
```

# Attachment of people to their opinion before discussion over time

## without any epsilon

```python
teams_attachment = gll.compute_all_teams_attachments(
    teams_data=teams_data,
    start_k=0,
    use_attachment_to_initial_opinion=False,
    eps=0)

plot_attachment_to_beforedisc_opinion(
    teams_attachment=teams_attachment, eps=0, fname='Figures/Attachment_to_beforedisc_opinion')
```

```python
all_aiis, all_nan_details_stats = get_all_aiis(teams_attachment)

print('len of all_aiis:', len(all_aiis))
all_aiis_with_no_nans = all_aiis[~np.isnan(all_aiis)]
print('len of all_aiis_with_no_nans:', len(all_aiis_with_no_nans))
all_aiis_with_no_nans_in_range = [x for x in all_aiis_with_no_nans if -1 <= x <= 1]
print('len of all_aiis_with_no_nans_in_range:', len(all_aiis_with_no_nans_in_range))
plt.hist(all_aiis_with_no_nans_in_range, bins=20)
plt.xlabel('Attachment to the before discussion opinion for every person')
plt.ylabel('Volume');
```

```python
print('{:.2f}% of all attachment to the initial opinions were one.'.format(
    100 * len(np.where(all_aiis == 1)[0]) / len(all_aiis)))

print('{:.2f}% of all attachment to the initial opinions were zero.'.format(
    100 * len(np.where(all_aiis == 0)[0]) / len(all_aiis)))

print('{:.2f}% of all attachment to the initial opinions were inside of range [0, 1] (including both 0 and 1).'.format(
    100 * len(np.where((all_aiis >= 0) & (all_aiis <= 1))[0]) / len(all_aiis)))

print('{:.2f}% of all attachment to the initial opinions were Nan (division by zero).'.format(
    100 * np.sum(np.isnan(all_aiis)) / len(all_aiis)))
```

## Testing to what extent $$a_{i, i} == 1 - w_{i, i} = w_{i, j}$$

```python
diffs = []
for team_id, team_data in teams_data.items():
    for task_name, team_this_task in team_data.items():
        aiis = teams_attachment[team_id][task_name]
        a11 = np.array(aiis['a11'])
        w12 = np.array(team_this_task['w12'])
        a22 = np.array(aiis['a22'])
        w21 = np.array(team_this_task['w21'])
        diffs.extend(abs(w21 - a22))
        diffs.extend(abs(w12 - a11))
diffs = np.array(diffs)
diffs = diffs[~np.isnan(diffs)]
plt.hist(diffs[diffs < 10])
```

```python
len(np.where(diffs < 0.7)[0]) / len(diffs)
```

```python
len(np.where(diffs == 0)[0]) / len(diffs)
```
