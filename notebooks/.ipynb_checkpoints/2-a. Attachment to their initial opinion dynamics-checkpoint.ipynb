# Helper functions

```python
def plot_attachment_to_initial_opinion(teams_attachment, eps, fname=''):
    teams_num = len(teams_attachment.keys())
    issues_num = 5
    fig, axes = plt.subplots(teams_num, issues_num, figsize=(issues_num*3, teams_num*3))
    for team_index, team_id in enumerate(sorted(teams_attachment.keys())):
        for issue_index, issue in enumerate(sorted(teams_attachment[team_id].keys())):
            a11 = teams_attachment[team_id][issue]['a11']
            a22 = teams_attachment[team_id][issue]['a22']
            # Plotting the level of attachment to the initial opinion dynamics.
            axes[team_index, issue_index].plot(a11, '-x')
            axes[team_index, issue_index].plot(a22, '-p')
            axes[team_index, issue_index].legend(['Person 1', 'Person 2'])
            if team_index == 0:
                axes[team_index, issue_index].set_title(issue)
            if team_index == teams_num - 1:
                axes[team_index, issue_index].set_xlabel('Time')
            if issue_index == 0:
                axes[team_index, issue_index].set_ylabel(
                    'Team {}\'s attachment'.format(team_id))
    if fname != '':
        fig.savefig(fname+'_eps{}.pdf'.format(eps), bbox_inches='tight')
        fig.savefig(fname+'_eps{}.png'.format(eps), bbox_inches='tight')
```

```python
nan_keys = ['0/0', 'n/0', 'xi[k]-xi[0]==0', 'wij[k]==0', 'xj[k]-xi[k]==0']

def get_all_aiis(teams_attachment):
    all_aiis = []
    nan_stats = defaultdict(int)
    for team_index, team_id in enumerate(teams_attachment.keys()):
        for issue_index, issue in enumerate(teams_attachment[team_id].keys()):
            a11 = teams_attachment[team_id][issue]['a11']
            a22 = teams_attachment[team_id][issue]['a22']
            all_aiis.extend(a11)
            all_aiis.extend(a22)
            
            a11_nan_details = teams_attachment[team_id][issue]['a11_nan_details']
            a22_nan_details = teams_attachment[team_id][issue]['a22_nan_details']
            
            for i, x in enumerate(a11):
                if len(a11_nan_details[i]) > 0 and not np.isnan(x):
                    print(team_id, issue, 1, i)
            for i, x in enumerate(a22):
                if len(a22_nan_details[i]) > 0 and not np.isnan(x):
                    print(team_id, issue, 2, i)
            
            for nan_key in nan_keys:
                for a11_nan_detail_item in a11_nan_details:
                    if nan_key in a11_nan_detail_item:
                         nan_stats[nan_key] += 1
                for a22_nan_detail_item in a22_nan_details:
                    if nan_key in a22_nan_detail_item:
                        nan_stats[nan_key] += 1
    return np.array(all_aiis), nan_stats
```

# Main  body

```python
n, _ = data.shape
team_ids = np.unique(data['Group'])
issues = ['asbestos', 'disaster', 'school', 'sports', 'surgery']
```

```python
consensus_opinions = []

consensuses = np.zeros(4)
teams_data = defaultdict(dict)
op_stats = defaultdict(int)
op_checks = 0
at_the_end_consensuses = defaultdict(list)
in_begining_consensuses = defaultdict(list)
for team_index, team_id in enumerate(team_ids):
    this_team_data = data[data['Group'] == team_id]
    for issue_index, issue in enumerate(issues):
        this_team_issue_data = this_team_data[this_team_data['Issue'] == issue]
        this_team_issue_data = this_team_issue_data.sort_values(
            by='Person')   # Just to make sure the first one is person 1.

        p1 = this_team_issue_data.iloc[0]
        x1 = [float(re.sub('\$|,', '', p1['Initial opinion'])),
              float(re.sub('\$|,', '', p1['Period1 opinion'])),
              float(re.sub('\$|,', '', p1['Period2 opinion'])),
              float(re.sub('\$|,', '', p1['Period3 opinion']))]
        w12 = [float(p1['Period1 wij']),
               float(p1['Period2 wij']),
               float(p1['Period3 wij'])]

        p2 = this_team_issue_data.iloc[1]
        x2 = [float(re.sub('\$|,', '', p2['Initial opinion'])),
              float(re.sub('\$|,', '', p2['Period1 opinion'])),
              float(re.sub('\$|,', '', p2['Period2 opinion'])),
              float(re.sub('\$|,', '', p2['Period3 opinion']))]
        w21 = [float(p2['Period1 wij']),
               float(p2['Period2 wij']),
               float(p2['Period3 wij'])]
        teams_data[team_id][issue] = {'x1': x1, 'x2': x2, 'w12': w12, 'w21': w21}

        # Getting the stats about how many times x[i]==x[0], x[i]==x[i-1], and x1[i]==x2[i] has happened.
        for i in range(1, 4):
            if x1[i] == x1[0]:
                op_stats['x[i]==x[0]'] += 1
            if x1[i] == x1[i-1]:
                op_stats['x[i]==x[i-1]'] += 1
            if x1[i] == x2[i]:
                op_stats['x1[i]==x2[i]'] += 1
            if x2[i] == x2[0]:
                op_stats['x[i]==x[0]'] += 1
            if x2[i] == x2[i-1]:
                op_stats['x[i]==x[i-1]'] += 1
        op_checks += 6
        
        # Consensus between two people for any issue at every round.
        this_team_consensus_opinions = []
        for i in range(4):
            if x1[i] == x2[i]:
                consensuses[i] += 1
                this_team_consensus_opinions.append(x1[i])
            else:
                this_team_consensus_opinions.append(0)
        consensus_opinions.append(this_team_consensus_opinions)

        # Consensus between two people at the end.
        if x1[3] == x2[3]:
            at_the_end_consensuses[issue].append(1)
        else:
            at_the_end_consensuses[issue].append(0)
        # Consensus between two people in the very first answer.
        if x1[0] == x2[0]:
            in_begining_consensuses[issue].append(1)
        else:
            in_begining_consensuses[issue].append(0)
```

# Prediction for the next timestamp using A, W and x in the current time

```python
def predict_FJ(A, W, x, x0=None):
#     if np.sum(np.isnan(A)) > 0 or np.sum(np.isnan(W)) > 0 or np.sum(np.isnan(x)) > 0:
#         print('ERRRRRRRRRRRRRROR')
    n, _ = W.shape
    if x0 is None:
        x0 = x
    for k in range(100):
#         x = np.matmul(np.matmul(A, W), x) + np.matmul((np.eye(n) - A), x0)
        x = A @ W @ x + (np.eye(n) - A) @ x0
    return x

def predict_Degroot(W, x):
    for k in range(100):
#         x = np.matmul(W, x)
        x = W @ x
    return x
```

```python
with Timer():
    predictions_nochange = defaultdict(list) 
    predictions_fj_init = defaultdict(list)
    predictions_fj_prev = defaultdict(list)
    predictions_degroot = defaultdict(list)
    for team_id, team_data in teams_data.items():
        for task_name, team_task_data in team_data.items():
            for i in range(2):
                a11 = teams_attachment[team_id][task_name]['a11'][i]
                a22 = teams_attachment[team_id][task_name]['a22'][i]

                # BASED ON Noah's recommendation.
                if np.isnan(a11):
                    a11 = 0
                if np.isnan(a22):
                    a22 = 0
                # BASED ON Noah's recommendation.

                w12 = team_task_data['w12'][i]
                w21 = team_task_data['w21'][i]
                x1s = team_task_data['x1']
                x2s = team_task_data['x2']
                A = np.array([[a11, 0], [0, a22]])
                W = np.array([[1-w12, w12], [w21, 1-w21]])
                x = np.array([x1s[i], x2s[i]])
                x0 = np.array([x1s[0], x2s[0]])
                true_next_x = np.array([x1s[i+1], x2s[i+1]])

                pred_next_x_fj_init = predict_FJ(A, W, x, x0)
                predictions_fj_init['pred'].extend(pred_next_x_fj_init)
                predictions_fj_init['true'].extend(true_next_x)

                pred_next_x_fj_prev = predict_FJ(A, W, x)
                predictions_fj_prev['pred'].extend(pred_next_x_fj_prev)
                predictions_fj_prev['true'].extend(true_next_x)

                pred_next_x_degroot = predict_Degroot(W, x)
                predictions_degroot['pred'].extend(pred_next_x_degroot)
                predictions_degroot['true'].extend(true_next_x)
                
                predictions_nochange['pred'].extend(x)
                predictions_nochange['true'].extend(true_next_x)
```
